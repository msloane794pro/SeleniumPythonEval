```python
#Imports
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Start the session
driver = webdriver.Chrome()

# Take action on the browser (Note, this URL won't work as it has been fixed to avoid sharing pirivate data.)
driver.get("https://app.powerbiXXX.XX/groups/d5c88659-e5d8-4a72-801b-9056521292d3/reports/06aa3e3e-7802-474a-a983-b7b49928f501/907bbe0ead769e6ebc1b?bookmarkGuid=1b80847497113b9509b5")

```

```python
numElements = 0
supersetAttributeList = set()

# Initialize the elementData dictionary
elementData = {}
row_number = 1

mat_list_elements = driver.find_elements(by=By.CLASS_NAME, value="mat-list-item")
for mat_list_element in mat_list_elements:
    mat_list_element.click()
    
    time.sleep(5)  # Wait for 5 seconds for the page to load
    
    # Retrieve and print information for various containers
    container_classes = [
        "visualContainer",
        "visualContainerHost",
        "filterContainer",
        "slicerContainer",
        "reportContainer",
        "pageContainer",
        "tileContainer",
        "dashboardContainer",
        "actionBarContainer",
        "navigationContainer"
    ]
    
    for container_class in container_classes:
        elements = driver.find_elements(by=By.CLASS_NAME, value=container_class)
        if len(elements) > 0:
            print(f'{mat_list_element.text}: {container_class} has {len(elements)} elements.')
            numElements = numElements + len(elements)
            
            # Find all the attributes available for each element in the list of elements
            for element in elements:
                attributes = driver.execute_script(
                    'var items = {}; for (index = 0; index < arguments[0].attributes.length; ++index) { items[arguments[0].attributes[index].name] = arguments[0].attributes[index].value }; return items;',
                    element)
                
                # Add to the elementData dictionary
                elementData[row_number] = {'Page': mat_list_element.text, 'Container': container_class, 'element-id': element.id, 'text': element.text, 'aria-role': element.aria_role}
                elementData[row_number].update(attributes)
                
                row_number += 1

print(f'Total number of elements observed: {numElements}')
driver.close()

# Create a Pandas dataframe from the elementData dictionary
df = pd.DataFrame.from_dict(elementData, orient='index')

# Remove columns from the Pandas dataframe where the column name starts with a "_" character
df = df.loc[:, ~df.columns.str.startswith('_')]

# Sort the columns from the Pandas dataframe in alphabetical order
df = df.reindex(sorted(df.columns), axis=1)

primaryColumns = ['Page', 'Container', 'aria-roledescription', 'aria-label', 'text']

# Move the columns for "Page" and "Container" to the front of the columns
columns = primaryColumns + [col for col in df.columns if col not in primaryColumns]
df = df[columns]

# Write the Pandas dataframe to an Excel file
df.to_excel("output.xlsx", index=False)

print("Data has been written to output.xlsx")
```
